```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Data:
    def __init__(self, filename) -> None:
        self.filename = filename
        self.data = self.read_data()
    
    def read_data(self):
        with open(self.filename, 'r') as f:
            data = f.read()
            print(data[0])
        return data
```

```python
file_name = 'datas/1234567890-100-1000-100-100.txt'
with open(file=file_name, mode='r') as file:
    lines = file.readlines()
    variable_names = lines[0]
    variables = lines[1]
    data_names = lines[2]
    data = lines[4:-1]
    time = lines[-1]


print(f'sedd = -{variables.split()[0]}')
print(f'mcs = {variables.split()[1]}')
print(f'temperatur= {variables.split()[2]}K')
print(f'CrAtoms = {variables.split()[3]}')
print(f'nx = {variables.split()[4]}')
print(f'ny = {variables.split()[5]}')


mcs = []
energy = []
magnetization = []
for i in range(len(data)):
    mcs.append(float(data[i].split()[0]))
    energy.append(float(data[i].split()[1]))
    magnetization.append(float(data[i].split()[2]))


```

```python
plt.plot(mcs, magnetization)
plt.xlabel('Monte Carlo Cycles')
plt.ylabel('Magnetization')
plt.title('Magnetization vs Monte Carlo Cycles')
plt.show()
```

```python
plt.plot(mcs, energy)
plt.xlabel('Monte Carlo Cycles')
plt.ylabel('Energy')
plt.title('Energy vs Monte Carlo Cycles')
plt.show()
```
